```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Administrator\Desktop\alex the da\warehouse_messy_data.csv'
                )
df
```

```python
df.columns
```

```python
df['Category'] = df['Category'].str.capitalize()
df
```

```python
# working on the quantity column
df['Quantity'] = df['Quantity'].replace('two hundred', 200)
df['Quantity'] = df['Quantity'].replace('NaN', np.nan)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
#working on price column
df['Price'] = df['Price'].replace('NaN', np.nan)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
#working on restocked column
df['Last Restocked'] = df['Last Restocked'].replace('NaN', np.nan)
```

```python
df
```

```python
# filling NaN value with default value or strategy
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

```python
# filling NaN values for categorical columns with placeholder
df['Category'] = df['Category'].fillna('Unknown Category')
df['Product Name'] = df['Product Name'].fillna('Unknown Product')
df['Warehouse'] = df['Warehouse'].fillna(df['Warehouse'].mode()[0])
df['Location'] = df['Location'].fillna('Unknown Location')
df['Supplier'] = df['Supplier'].fillna(df['Supplier'].mode()[0])
df['Status'] = df['Status'].fillna(df['Status'].mode()[0])
df['Last Restocked'] = df['Last Restocked'].fillna(pd.to_datetime('today'))
```

```python
df.to_csv(r'C:\Users\Administrator\Desktop\alex the da\warehouse_messy_data.csv'
                )
```
